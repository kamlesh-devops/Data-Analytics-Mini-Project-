```python
df = pd.read_csv(r"C:\Users\jkaml\Downloads\unprocessed dataset (amazon sales).csv")
df.columns
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import rand_score, silhouette_score

df['OrderDate_num'] = pd.to_datetime(df['OrderDate']).map(pd.Timestamp.toordinal)

X = pd.get_dummies(df[['OrderDate_num', 'Category', 'TotalAmount', 'PaymentMethod']])
y = df['OrderStatus'].astype(str)

results = []
for k in [3, 5, 7]:
    # Initialize and fit the model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X)
    inertia = kmeans.inertia_
    sil_score = silhouette_score(X, clusters)
    ri_score = rand_score(y, clusters) # Compares clusters against true OrderStatus
    
    results.append({
        'K': k,
        'Inertia': inertia,
        'Silhouette Score': sil_score,
        'Rand Index': ri_score
    })

performance_df = pd.DataFrame(results)
print(performance_df)
```
